```
from keras.models import load_model

classifier = load_model("/content/drive/My Drive/Colab Notebooks/CNNs/models/mnist_digit_recognition_5_epoch_cnn.h5")
```

```
model = classifier
```

```
#for visualization, we use a separate library - VIS
#!pip install -I scipy==1.2.*
from vis.visualization import visualize_activation
from vis.utils import utils

from keras import activations
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (18,6)

layer_idx = -1 #since we are visualising prediction layer (last)

model.layers[layer_idx].activation = activations.linear #we replace the softmax layer by linear

model = utils.apply_modifications(model)

#we maximize the output layer
filter_idx = 0 #the prediction
img = visualize_activation(model,layer_idx,filter_indices = filter_idx)

plt.imshow(img[...,0])
```

```
img = visualize_activation(model,layer_idx,filter_indices = filter_idx, input_range=(0.,1.), verbose =True)

plt.imshow(img[...,0])
```

```
import numpy as np

for output_idx in list(range(10)):
  img = visualize_activation(model,layer_idx,filter_indices = output_idx, input_range=(0.,1.))
  plt.figure()
  plt.title('Network perception of {}'.format(output_idx))

  plt.imshow(img[...,0])
```
