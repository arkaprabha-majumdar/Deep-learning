```
from keras.preprocessing.image import ImageDataGenerator

#let's import a random image
import urllib
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

req = urllib.request.urlopen("http://2.bp.blogspot.com/-k1LTVxtn0Ls/UA5xzZv7o9I/AAAAAAAANNE/7Wg33Qnv-Aw/s1600/Lion5+-+Copy.jpgg")
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.resize(cv2.imdecode(arr, -1),(180,150))
cv2_imshow(img)
```

```
import matplotlib.pyplot as plt

datagen = ImageDataGenerator(rotation_range=60)

i=0
for batch in datagen.flow(img.reshape((1,) + img.shape)):
  if i<9:
    print(i)
    plt.subplot(330+1+i)
    plt.imshow(batch.reshape(150,180,3).astype(int))
  else:
    break
  i+=1
plt.show()
```

```
#Similarly, we can apply other affine and non-affine transformations

datagen = ImageDataGenerator(rotation_range = 50,
                             shear_range = 0.5,
                             zoom_range = 0.5)

i=0
plt.figure(figsize=(20,20))
for batch in datagen.flow(img.reshape((1,) + img.shape)):
  if i<25:
    plt.subplot(5,5,1+i)
    plt.title('tiger {}'.format(i))
    plt.imshow(batch.reshape(150,180,3).astype(int))
  else:
    break
  i+=1
plt.show()
```
