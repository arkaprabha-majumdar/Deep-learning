# #1. Face Detection

```python
import cv2
import numpy as np
```

```python
#initialize the haar frontal face cascade
face_classifier = cv2.CascadeClassifier("/home/arkaprabham/Desktop/Deep_Learning_Udemy/my_project_env/lib/python3.8/site-packages/cv2/data/haarcascade_frontalface_default.xml")

image = cv2.imread("/home/arkaprabham/Desktop/Deep_Learning_Udemy/img/Animals/humans/humans.jpg")

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

faces = face_classifier.detectMultiScale(gray,1.3,5)

if faces.size:
    for (x,y,w,h) in faces:
        cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)
        cv2.imshow("Faces detected",image)
        cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
#initialize the haar frontal face cascade
face_classifier = cv2.CascadeClassifier("/home/arkaprabham/Desktop/Deep_Learning_Udemy/my_project_env/lib/python3.8/site-packages/cv2/data/haarcascade_frontalface_default.xml")

#Let us put the above code into a function and we'll do something nice !
def face_detect(image):
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    
    faces = face_classifier.detectMultiScale(gray,1.3,5)

    if list(faces):
        for (x,y,w,h) in faces:
            cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)
    else:
        return image
    return image


#remember how we accessed the webcam in our initial projects?
cap = cv2.VideoCapture(0)
i=0
while True:
    print(i)
    _,frame=cap.read()
    cv2.imshow("Face detector",face_detect(frame))
    i+=1
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
#Excuse the frame number being printed, but I like to see my code running each loop
```

# #2. Combining eye detection with face directly for webcam


```python
#initialize the haar frontal face classifier AND eye classifier
face_classifier = cv2.CascadeClassifier("/home/arkaprabham/Desktop/Deep_Learning_Udemy/my_project_env/lib/python3.8/site-packages/cv2/data/haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier("/home/arkaprabham/Desktop/Deep_Learning_Udemy/my_project_env/lib/python3.8/site-packages/cv2/data/haarcascade_eye.xml")


#I copy pasted the code above and made modifications...try figuring it out yourself. NOT THAT HARD!
def face_eye_detect(image):
    blur = cv2.GaussianBlur(image,(3,3),1)
    gray = cv2.cvtColor(blur,cv2.COLOR_BGR2GRAY)
    
    faces = face_classifier.detectMultiScale(gray,1.3,5)
    
    if list(faces):
        for (x,y,w,h) in faces:
            cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)
            roi_gray=gray[y:y+h,x:x+w]
            roi_color=image[y:y+h,x:x+w]
            print(roi_gray)
            if list(roi_gray):
                eyes = eye_classifier.detectMultiScale(roi_gray)
                if list(eyes):
                    print("Eyes")
                    for (ex,ey,ew,eh) in eyes:
                        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
                else:
                    print("No eyes")
    else:
        return image
    return image


#remember how we accessed the webcam in our initial projects?
cap = cv2.VideoCapture(0)
i=0
while True:
    print(i)
    _,frame=cap.read()
    cv2.imshow("Face detector",face_eye_detect(frame))
    i+=1
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
#Excuse the frame number being printed, but I like to see my code running each loop
```
